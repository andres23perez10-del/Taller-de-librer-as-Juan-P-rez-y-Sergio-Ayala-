```python
print("Registro de Activos De Informacion RTCV")
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv ('Registro de activos RTCV.csv')
df.columns = df.columns.str.strip()
display(df.head())

plt.figure(figsize=(10, 5))
top_formatos = df['formato'].value_counts().head(10)
plt.bar(top_formatos.index, top_formatos.values, color='#2ca02c', edgecolor='black', width=0.5)
plt.title('Top Formatos de Activos de Información más Utilizados en RTVC', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Tipo de Formato / Naturaleza del Activo', fontsize=11)
plt.ylabel('Cantidad de Registros', fontsize=11)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
plt.show()


print("Tamaño de Bytes de datos del portal nacional de datos abiertos")
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv ('Tamaño Bytes.csv')
df.columns = df.columns.str.strip()
display(df.head())
plt.figure(figsize=(10, 5))
muestra_distribucion = df.sort_values(by='uid_size_gigabytes', ascending=False).iloc[10:20]
plt.bar(muestra_distribucion['uid'], muestra_distribucion['uid_size_gigabytes'], color='#1f77b4', edgecolor='black', width=0.5)
plt.title('Distribución de Tamaño en Gigabytes (Datasets del Puesto 11 al 20)', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Identificador del Dataset (UID)', fontsize=11)
plt.ylabel('Tamaño en Gigabytes (GB)', fontsize=11)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()


print("Reporte de Errores MinTic")
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv ('Reporte de Errores.csv')
df.columns = df.columns.str.strip()
display(df.head())
plt.figure(figsize=(10, 5))
top_errores = df['Error'].value_counts().head(10)
plt.bar(top_errores.index, top_errores.values, color= '#1f77b4', edgecolor='black', width=0.5)

plt.title(' Errores más Frecuentes en el Portal MinTIC', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Código de Error Detectado', fontsize=11)
plt.ylabel('Cantidad de Reportes', fontsize=11)

plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)


```
